### Analysis

- Capomulin and  Ramicane seem to be the most effective treatments based on the data. They result in lower tumor volumes on average and do not have any outliers. This is important because it helps to show that, at least with this data, that the test results of these trials are more consitant.
- There does seem to be a positave corrilation between weight and tumor size. This could be important because this question can now be asked: If a mouse loses weight, will this have any effect on the tumor size and/or effectiveness of the treatment? Because there is a corrilation, finding the reason behind it is important for the study.
-Infubinol seemed to be the least effective regimen. It had larger average tumor volume and also had outliers which means that the effectiveness is less consistant when compared with the others.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_mouse_df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='outer')

# Display the data table for preview

merged_mouse_df.head()
```

```python
# Checking the number of mice.
mouse_count = merged_mouse_df['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_id = merged_mouse_df.loc[merged_mouse_df.duplicated(subset=['Mouse ID', "Timepoint"]), "Mouse ID"].unique()
duplicate_id
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicatte_mouse_data = merged_mouse_df.loc[merged_mouse_df['Mouse ID'] == "g989"] 
duplicatte_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merged_mouse_df[merged_mouse_df["Mouse ID"].isin(duplicate_id) == False] 
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_length = clean_df["Mouse ID"].nunique()
clean_mouse_length
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_mouse = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()

median_mouse = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()

variance_mouse = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()

std_mouse = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()

sem_mouse = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
new_mouse_df = pd.DataFrame({
    "Mean Tumor Volume":mean_mouse,
    "Median Tumor Volume": median_mouse,
    "Tumor Volume Variance": variance_mouse,
    "Tumor Volume Std. Dev.": std_mouse,
    "Tumor Volume Std. Err.": sem_mouse})
new_mouse_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
optional_df = clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
# Using the aggregation method, produce the same summary statistics in a single line
optional_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tumor = clean_df.groupby(["Mouse ID"])["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = final_tumor.merge(clean_df, on=["Mouse ID", "Timepoint"], how="left")
merged_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", 'Ceftamin'] 

# Create empty list to fill with tumor vol data (for plotting)

tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatments:
    

    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_final_vol = merged_data.loc[merged_data["Drug Regimen"] == treatment, 'Tumor Volume (mm3)']
    # add subset
    tumor_vol_list.append(tumor_final_vol)

    # Determine outliers using upper and lower bounds
    quart_range = tumor_final_vol.quantile([0.25, 0.5, 0.75])

    lower_quart = quart_range[0.25]
    upper_quart = quart_range[0.75]

    iqr = upper_quart - lower_quart
    low_bound = lower_quart - (1.5 * iqr)
    up_bound = upper_quart + (1.5 * iqr)

    possible_outliers = tumor_final_vol[(tumor_final_vol < low_bound) | (tumor_final_vol > up_bound)]

    print(f"{treatment} Possible Outliers:")
    if possible_outliers.empty:
        print("  No outliers detected")
    else:
        for outlier, value in possible_outliers.items():
            print(f"  Mouse ID {outlier}: Tumor Volume = {value}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
treatments = ["Capomulin", "Ramicane", "Infubinol", 'Ceftamin'] 
tumor_clean_df = merged_data.loc[merged_data["Drug Regimen"].isin(treatments)]

volume_regimen_box = tumor_clean_df.boxplot(column="Tumor Volume (mm3)", by="Drug Regimen",  grid=False)

volume_regimen_box.set_ylabel("Tumor Volume (mm3)")
volume_regimen_box.set_title("Tumor Volume by Treatment")


plt.show()
```

## Line and Scatter Plots

```python
#Create line
capomulin_df = clean_df[clean_df["Drug Regimen"] == "Capomulin"]


mouse_id = capomulin_df["Mouse ID"].unique()[0] 


mouse_data = capomulin_df[capomulin_df["Mouse ID"] == mouse_id]

plt.plot(mouse_data["Timepoint"], mouse_data["Tumor Volume (mm3)"], linestyle='-', color='blue')
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Tumor Volume vs. Timepoint for Mouse ID {mouse_id} (Capomulin Treatment)")

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = clean_df[clean_df["Drug Regimen"] == "Capomulin"]
avg_tum_vol = capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().reset_index()


cap_weight = capomulin_df[['Mouse ID', 'Weight (g)']].drop_duplicates()
cap_merged = avg_tum_vol.merge(cap_weight, on="Mouse ID")


plt.scatter(cap_merged["Weight (g)"], cap_merged["Tumor Volume (mm3)"], marker='o', color='blue')
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume for Capomulin Treatment")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = cap_merged["Weight (g)"]
y_values = cap_merged["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.annotate(line_eq,(20,15),fontsize=15,color="red")
plt.title(f"Mouse Weight vs. Average Tumor Volume for Capomulin Treatment")

# Print r^2 value
print(f"The correlation between mouse weight and the average tumor volume is: {rvalue**2}")

# Show plot
plt.show()
```
